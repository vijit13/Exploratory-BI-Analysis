# Time Series Analysis

```python
df=merged[['customer_id','start_visit', 'approved_visit', 'complete_visit', 'start_visit_localtime', 'complete_visit_localtime']]
df
```

```python
df.start_visit.dt.time.unique()
```

# Plot of the most popular purchase times based on the start time of the visit

```python
plt.hist(df.start_visit.dt.time.unique())
```

### From the plot it is clearly visible that the most popular time of purchases is between 5 PM to around 10:30 PM 

# Plot showing the dayofweek when the maximum number of purchases are made.

```python
plt.hist(df.start_visit.dt.dayofweek.values)
```

### The above plot shows that the customers generally tend to make most of their purchases on the weekends.

# Plot showing the overall trend of sales for the year 2017

```python
plt.hist(df.start_visit.dt.year.values)
```

```python
plt.hist(df.start_visit.dt.month.values)
```

## For the above plot we can infer that the sales have decreased drastically from February onwards.Maximum sales are in the months of January and February.
## The data is provided only for first 4 months in 2017 and Dec 2016.
